```python
import pandas as pd
import numpy as np
```

```python
reviewer_df = pd.read_csv("../../../Datasource/bgg-19m-reviews.csv")
print(reviewer_df)
```

```python
reviewer_df
```

```python
rating_bg = reviewer_df.groupby(["ID","name"])["rating"].mean()
rating_bg = pd.DataFrame(rating_bg.sort_values(ascending=False))
rating_bg = rating_bg.reset_index()
```

```python
rating_bg =rating_bg.rename(columns={"ID" : "boardgame_id", "name" : "boardgame_name"})
rating_bg
```

```python
bg_info = pd.read_csv("../../../Dataset/games_detailed_info.csv")
bg_info
```

```python
bg_info.columns
```

```python
cols = ['id', 'image', 'primary',
       'alternate', 'description', 'yearpublished', 'minplayers', 'maxplayers', 'playingtime', 'minplaytime',
       'maxplaytime', 'Board Game Rank']
bg_info = bg_info[cols]
```

```python
bg_info = bg_info.rename(columns={"id": "boardgame_id", "primary" : "boardgame_name", "alternate" : "name_2"})
bg_info
    
```

```python
data  = rating_bg.merge(bg_info, on=["boardgame_id", "boardgame_name"])
data
```

```python
data.to_csv("../artifacts/data_web.csv", index = False)
```
